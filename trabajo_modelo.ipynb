```python
#Tarelo cerna ramon antonio y armenta lopez eduardo
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import LinearSVC
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Cargar datos con Pandas
data = pd.read_csv('australian.dat', sep=' ', header=None)
X = data.iloc[:, :-1].values
y = data.iloc[:, -1].values
```

```python
# Modelo base para Random Forest
scores_base_rf = []
for i in range(10):
    clf = RandomForestClassifier(random_state=42)
    clf.fit(X, y)
    scores_base_rf.append(clf.score(X, y))
```

```python
# Modelo con hiperparámetros para Random Forest
clf = RandomForestClassifier(n_estimators=100, criterion="gini", max_depth=None, min_samples_split=2, min_samples_leaf=1, max_features=12, random_state=42)
scores_hiper_rf = []
for i in range(10):
    clf.fit(X, y)
    scores_hiper_rf.append(clf.score(X, y))

```

```python
# Boxplot para Random Forest
fig, ax = plt.subplots()
bp_dict = ax.boxplot([scores_base_rf, scores_hiper_rf], labels=['Base', 'Hiperparámetros'])
ax.set_title('Random Forest')
plt.show()
```

```python
# Modelo base para Linear Support Vector Classifier
scores_base_svc = []
for i in range(10):
    clf = LinearSVC(random_state=42)
    clf.fit(X, y)
    scores_base_svc.append(clf.score(X, y))
```

```python
# Modelo con hiperparámetros para Linear Support Vector Classifier
clf = LinearSVC(C=0.1, penalty='l1', dual=False, random_state=42)
scores_hiper_svc = []
for i in range(10):
    clf.fit(X, y)
    scores_hiper_svc.append(clf.score(X, y))
```

```python
# Boxplot para Linear Support Vector Classifier
fig, ax = plt.subplots()
bp_dict = ax.boxplot([scores_base_svc, scores_hiper_svc], labels=['Base', 'Hiperparámetros'])
ax.set_title('Linear Support Vector Classifier')
plt.show()
```

```python
# Comparación de los promedios de scores
if np.mean(scores_base_rf) > np.mean(scores_base_svc) and np.mean(scores_hiper_rf) > np.mean(scores_hiper_svc):
    print("El modelo de Random Forest es mejor")
else:
    print("El modelo de Linear Support Vector Classifier es mejor")
```
